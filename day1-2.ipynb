# Defining persona : History vs Math teacher (routing prompt)

```python
questions = [
    {
        "type": "history",
        "question": "Who was the first President of the United States?"
    },
    {
        "type": "history",
        "question": "In what year did the Titanic sink?"
    },
    {
        "type": "history",
        "question": "What was the primary cause of the American Civil War?"
    },
    {
        "type": "history",
        "question": "Who was the leader of the Soviet Union during World War II?"
    },
    {
        "type": "history",
        "question": "Which empire was known as the 'Empire on which the sun never sets'?"
    },
    {
        "type": "history",
        "question": "What impact did the Industrial Revolution have on society in the 19th century?"
    },
    {
        "type": "history",
        "question": "Why did the Roman Empire fall?"
    },
    {
        "type": "history",
        "question": "What were the main factors leading to the French Revolution?"
    },
    {
        "type": "history",
        "question": "Describe the significance of the Magna Carta in medieval England."
    },
    {
        "type": "math",
        "question": "What are the fundamental principles of calculus?"
    },
    {
        "type": "math",
        "question": "Explain the concept of probability in statistics."
    },
    {
        "type": "math",
        "question": "What is the Fibonacci sequence and its mathematical properties?"
    },
    {
        "type": "math",
        "question": "Discuss the applications of linear algebra in computer science."
    },
    {
        "type": "math",
        "question": "How is trigonometry used in real-world applications?"
    },
    {
        "type": "none",
        "question": "What are the key elements of Greek mythology?"
    },
    {
        "type": "none",
        "question": "What is the role of ethics in philosophy?"
    },
    {
        "type": "none",
        "question": "Explain the concept of existentialism in modern philosophy."
    },
    {
        "type": "none",
        "question": "Describe the characteristics of Renaissance art."
    },
    {
        "type": "none",
        "question": "What are the main themes in Shakespearean literature?"
    }
]
```

```python
import requests

url = "http://localhost:11434/api/generate"

prompt_template = """
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Determines if the user request is suitable for a math teacher persona or a history teacher persona.
Follow these steps:
1. Identify the nature of the user request. Determine if the request involves math-related concepts or historical events.
2. Consider the context of the request. Analyze if the content aligns with mathematical problem-solving or historical analysis.
3. Imagine two different experts. Picture a math teacher and a history teacher providing responses to the request based on their expertise.
4. Evaluate the request. Decide which teacher persona (math or history) is appropriate to address the user request.
5. Answer in the format "#Answer: math teacher" or "#Answer: history teacher" and "#Answer: none" if it does not belong to either.<|eot_id|>
<|start_header_id|>user<|end_header_id|>

{}<|eot_id|>
<|start_header_id|>assistant<|end_header_id|>
"""

correct_answers = 0
total_questions = len(questions)

for question in questions:
    full_prompt = prompt_template.format(question["question"])

    payload = {
        "model": "llama3",
        "prompt": full_prompt,
        "stream": False
    }

    response = requests.post(url, json=payload)

    if response.status_code == 200:
        api_response = response.json()
        if "response" in api_response:
            generated_answer = api_response["response"]
            # print(api_response["response"])

            if (question["type"] == "history" and "history" in generated_answer.lower()) or \
               (question["type"] == "math" and "math" in generated_answer.lower()) or \
               (question["type"] == "none" and "none" in generated_answer.lower()):
                correct_answers += 1
            else:
                print(f"Generated answer for '{question['question']}': {generated_answer} - Expected type: {question['type']}")

    else:
        print(f"Request failed with status code {response.status_code}")
        print("Error message:", response.text)

# Calculate the final score
score = correct_answers / total_questions * 100

# Print the results
print(f"\nTotal questions: {total_questions}")
print(f"Correct answers: {correct_answers}")
print(f"Final score: {score:.2f}%")
```

# Algorithm prompt

```python
test_cases = [{
    "input": [88, 1, 2, 9, 55],
    "output": [1, 2, 9, 55, 88]
},
{
    "input": [100, 3, 45, 2, 7, 18, 1, 99],
    "output": [1, 2, 3, 7, 18, 45, 99, 100]
},
{
    "input": [5, 4, 2, 1, 3],
    "output": [1, 2, 3, 4, 5]
}]
```

```python
import re

correct_answers = 0
total_questions = len(test_cases)

prompt_template = """
<|begin_of_text|><|start_header_id|>system<|end_header_id|>
    
You will receive a list and sorting m numbers among natural numbers from 1 to n.
Please sort the list in ASC order in below format:
#Answer [list of numbers]
<|eot_id|>
<|start_header_id|>user<|end_header_id|>

{}<|eot_id|><|start_header_id|>assistant<|end_header_id|>
"""

for case in test_cases:
    full_prompt = prompt_template.format(case["input"])
    
    payload = {
            "model": "llama3",
            "prompt": full_prompt,
            "stream": False
        }
    
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        api_response = response.json()
        if "response" in api_response:
            generated_answer = api_response["response"]
            # print(api_response["response"])
            numbers = re.findall(r'\d+', generated_answer)

            numbers = list(map(int, numbers))

            if numbers == case["output"]:
                correct_answers += 1
            else:
                print(f"Generated answer for '{case['input']}': {generated_answer} - Expected answer: {case['output']}")

    else:
        print(f"Request failed with status code {response.status_code}")
        print("Error message:", response.text)

score = correct_answers / total_questions * 100

print(f"\nTotal questions: {total_questions}")
print(f"Correct answers: {correct_answers}")
print(f"Final score: {score:.2f}%")

```
